```python
# Source usage:
weather_url = "https://www.timeanddate.com/weather/usa/new-york/historic"
weather_scraper = WeatherScraper(weather_url)
weather_data = weather_scraper.scrape_weather_data()

if weather_data is not None:
    print(weather_data)
# Convert 'Time' column to a datetime object for time-based analysis
weather_data['Time'] = pd.to_datetime(weather_data['Time'], errors='coerce')

# Extracting temperature data for analysis
temperature_data = weather_data['Temp'].str.replace('°F', '').astype(float)

# Calculate maximum and minimum temperatures
max_temp = temperature_data.max()
min_temp = temperature_data.min()

# Summary statistics of temperature
temp_summary = temperature_data.describe()

```

1. Temperature trends Analysis: 

```python
# Plotting temperature trends over the 24-hour period
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(weather_data['Time'], temperature_data, marker='o')
plt.xlabel('Time')
plt.ylabel('Temperature (°F)')
plt.title('Temperature Trends in the Past 24 Hours')
plt.grid(True)
plt.show()
```

2. Humidity Variation Analysis:

```python
import pandas as pd
import matplotlib.pyplot as plt

# Convert 'Time' column to a datetime object for time-based analysis
weather_data['Time'] = pd.to_datetime(weather_data['Time'], errors='coerce')

# Extracting humidity data for analysis
humidity_data = weather_data['Humidity'].str.replace('%', '').astype(float)

# Humidity Variation Analysis
plt.figure(figsize=(10, 6))
plt.plot(weather_data['Time'], humidity_data, marker='o', linestyle='-', color='blue')
plt.title('Humidity Variation Throughout the Day')
plt.xlabel('Time')
plt.ylabel('Humidity (%)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

3. Barometric Pressure Variation Analysis:

```python
# Extracting barometric pressure data for analysis
barometer_data = weather_data['Barometer'].str.replace('"Hg', '').astype(float)

# Barometric Pressure Variation Analysis
plt.figure(figsize=(10, 6))
plt.plot(weather_data['Time'], barometer_data, marker='o', linestyle='-', color='green')
plt.title('Barometric Pressure Variation Throughout the Day')
plt.xlabel('Time')
plt.ylabel('Barometric Pressure (inches)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

4. Visibility Trends Analysis:

```python
# Extracting visibility data for analysis
visibility_data = weather_data['Visibility'].str.replace('mi', '').astype(float)

# Visibility Trends Analysis
plt.figure(figsize=(10, 6))
plt.plot(weather_data['Time'], visibility_data, marker='o', linestyle='-', color='orange')
plt.title('Visibility Trends Throughout the Day')
plt.xlabel('Time')
plt.ylabel('Visibility (miles)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

5. Time-based Analysis - Comparing Temperature and Weather Conditions:


```python
# Extracting temperature data for analysis
temperature_data = weather_data['Temp'].str.replace('°F', '').astype(float)

# Time-based Analysis - Comparing Temperature and Weather Conditions
plt.figure(figsize=(10, 6))
plt.plot(weather_data['Time'], temperature_data, marker='o', linestyle='-', color='red', label='Temperature')
plt.plot(weather_data['Time'], humidity_data, marker='o', linestyle='-', color='blue', label='Humidity')
plt.plot(weather_data['Time'], barometer_data, marker='o', linestyle='-', color='green', label='Barometer')
plt.title('Time-based Analysis of Temperature, Humidity, and Barometric Pressure')
plt.xlabel('Time')
plt.ylabel('Values')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
